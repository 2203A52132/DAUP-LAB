```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('insurance.csv')
```

```python
ages_vector = data['age']
smoker_vector = data['smoker']
gender_factor = pd.Categorical(data['sex'])
matrix = data[['age', 'bmi', 'expenses']].to_numpy()
filtered_df = data[['age', 'smoker', 'expenses']]
```

```python
plt.figure(figsize=(8, 6))
colors = data['smoker'].map({'yes': 'red', 'no': 'blue'})
plt.scatter(data['age'], data['smoker'], c=colors, alpha=0.6)
plt.title('Scatter Plot: Age vs Smoking')
plt.xlabel('Age')
plt.ylabel('Smoking Status')
plt.grid(alpha=0.3)
plt.legend(['Smokers (red)', 'Non-Smokers (blue)'], loc='upper right')
plt.show()
```

```python
filtered_data = data[(data['age'] >= 20) & (data['age'] <= 40)]
plt.figure(figsize=(8, 6))
colors_filtered = filtered_data['smoker'].map({'yes': 'red', 'no': 'blue'})
plt.scatter(filtered_data['age'], filtered_data['smoker'], c=colors_filtered, alpha=0.6)
plt.title('Scatter Plot: Age (20-40) vs Smoking')
plt.xlabel('Age')
plt.ylabel('Smoking Status')
plt.grid(alpha=0.3)
plt.legend(['Smokers (red)', 'Non-Smokers (blue)'], loc='upper right')
plt.show()
```

```python
smoker_expenses = data.groupby('smoker')['expenses'].mean()
smoker_expenses.plot(kind='bar', color=['blue', 'red'], figsize=(8, 6))
plt.title('Average Expenses: Smokers vs Non-Smokers')
plt.ylabel('Average Expenses')
plt.xlabel('Smoking Status')
plt.xticks(rotation=0)
plt.show()
```

```python
gender_smoker = data[data['smoker'] == 'yes']['sex'].value_counts()
gender_smoker.plot(kind='bar', color=['orange', 'purple'], figsize=(8, 6))
plt.title('Number of Smokers by Gender')
plt.ylabel('Count')
plt.xlabel('Gender')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['age'], data['expenses'], alpha=0.6, c='green')
plt.title('Scatter Plot: Age vs Expenses')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.grid(alpha=0.3)
plt.show()
```
